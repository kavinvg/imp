```r
install.packages("igraph")
library(igraph)
```

```r

g <- make_ring(10)

g <- add_edges(g, c(1, 5, 2, 6, 3, 7, 4, 8))
```

```r
plot(g, vertex.label = V(g)$name, main = "Simple Graph")

community_louvain <- cluster_louvain(g)

print(community_louvain)

V(g)$color <- membership(community_louvain)

plot(g, vertex.label = V(g)$name, vertex.color = V(g)$color,
     main = "Community Detection using Louvain Method",
     vertex.size = 25, edge.arrow.size = 0.5)
```

```r
community_fast_greedy <- cluster_fast_greedy(g)
plot(community_fast_greedy, g, main = "Community Detection using Fast-Greedy")
community_infomap <- cluster_infomap(g)
plot(community_infomap, g, main = "Community Detection using Infomap")
community_walktrap <- cluster_walktrap(g)
plot(community_walktrap, g, main = "Community Detection using Walktrap")
```
