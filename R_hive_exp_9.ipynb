```r
# Install packages
install.packages("ggplot2")
install.packages("dplyr")
install.packages("plotly")
install.packages("htmlwidgets")
library(plotly)
library(htmlwidgets)
library(ggplot2)
library(dplyr)
library(plotly)
```

```r
data <- read.csv("/content/sample_data/california_housing_train.csv")

head(data)
```

```r
# Create a scatter plot of median_income vs median_house_value
ggplot(data, aes(x = median_income, y = median_house_value)) +
  geom_point() +
  theme_minimal() +
  labs(title = "Scatter Plot of Median Income vs Median House Value",
       x = "Median Income",
       y = "Median House Value")
```
